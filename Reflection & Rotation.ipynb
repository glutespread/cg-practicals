```python
import numpy as np
import math
import matplotlib.pyplot as plt

class Triangle:
    
    def __init__(self,p1,p2,p3):
        self.p1 = p1
        self.p2 = p2
        self.p3 = p3
        self.x = [self.p1[0],self.p2[0],self.p3[0],self.p1[0]]
        self.y = [self.p1[1],self.p2[1],self.p3[1],self.p1[1]]
        
    def reflect_x(self):
        refx_mat = np.array([[1,0],[0,-1]])
        
        self.refx_p1 = np.matmul(p1,refx_mat)
        self.refx_p2 = np.matmul(p2,refx_mat)
        self.refx_p3 = np.matmul(p3,refx_mat)
        
        refx_x =[self.refx_p1[0],self.refx_p2[0],self.refx_p3[0],self.refx_p1[0]]
        refx_y = [self.refx_p1[1],self.refx_p2[1],self.refx_p3[1],self.refx_p1[1]] 
        plt.plot(self.x,self.y)
        plt.plot(refx_x,refx_y)
        plt.show()
    
    def reflect_y(self):
        refy_mat = np.array([[-1,0],[0,1]])
        
        self.refy_p1 = np.matmul(p1,refy_mat)
        self.refy_p2 = np.matmul(p2,refy_mat)
        self.refy_p3 = np.matmul(p3,refy_mat)
        
        refy_x =[self.refy_p1[0],self.refy_p2[0],self.refy_p3[0],self.refy_p1[0]]
        refy_y = [self.refy_p1[1],self.refy_p2[1],self.refy_p3[1],self.refy_p1[1]] 
        plt.plot(self.x,self.y)
        plt.plot(refy_x,refy_y)
        plt.show()
   

    def rotate(self,theta):
        theta = math.radians(theta)
        rot_mat = np.array([[math.cos(theta),(math.sin(theta))],[-(math.sin(theta)),math.cos(theta)]])
        
        self.rot_p1 = np.matmul(self.p1,rot_mat)
        self.rot_p2 = np.matmul(self.p2,rot_mat)
        self.rot_p3 = np.matmul(self.p3,rot_mat)
        
        rot_x = [self.rot_p1[0],self.rot_p2[0],self.rot_p3[0],self.rot_p1[0]]
        rot_y = [self.rot_p1[1],self.rot_p2[1],self.rot_p3[1],self.rot_p1[1]]
        plt.plot(self.x,self.y)
        plt.plot(rot_x,rot_y)
        plt.show()


p1 = np.array([1,1])
p2 = np.array([2,2])
p3 = np.array([3,1])
myTri = Triangle(p1,p2,p3)  
myTri.reflect_x()
myTri.reflect_y()
myTri.rotate(180)

        
        
        
        
        
```
