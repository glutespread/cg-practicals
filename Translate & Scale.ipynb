```python
import numpy as np
import math
import matplotlib.pyplot as plt

class Triangle:
    
    def __init__(self,p1,p2,p3):
        self.p1 = p1
        self.p2 = p2
        self.p3 = p3
        self.x = [self.p1[0],self.p2[0],self.p3[0],self.p1[0]]
        self.y = [self.p1[1],self.p2[1],self.p3[1],self.p1[1]]


    def translate(self,Tx,Ty):
        trans_x = [i+Tx for i in self.x]
        trans_y = [i+Ty for i in self.y]
        plt.plot(self.x,self.y)
        plt.plot(trans_x,trans_y)
        plt.show()
        
    def scale(self,sf):
        scaled_x = [i*sf for i in self.x]
        scaled_y = [i*sf for i in self.y]
        
        plt.plot(self.x,self.y)
        plt.plot(scaled_x,scaled_y)
        plt.show()

p1 = np.array([1,1])
p2 = np.array([2,2])
p3 = np.array([3,1])
myTri = Triangle(p1,p2,p3) 
myTri.translate(3,5)
myTri.scale(10)
```
